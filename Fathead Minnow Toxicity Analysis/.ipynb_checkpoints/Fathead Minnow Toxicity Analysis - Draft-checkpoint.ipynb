```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df['NdsCH'] = df['Chemicals'].apply(lambda x: float(x.split(";")[3]))
```

```python
df['NdssC'] = df['Chemicals'].apply(lambda x: float(x.split(";")[4]))
```

```python
df['MLOGP'] = df['Chemicals'].apply(lambda x: float(x.split(";")[5]))
```

```python
# No clear linear relationship between one quantity with the response variate LC50.
# Notice that MLOGP has a very close-to-linear correlation with LC50. 
# Attempt to use SVR, regression in Scikit_learn. If does not work, apply ANN from Tensorflow.
```

```python
# First, use SVR.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('LC50 [-LOG(mol/L)]', axis = 1)
y = df['LC50 [-LOG(mol/L)]']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear')
```

```python
svr.fit(X_train, y_train)
```

```python
predictions = svr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print('MAE:', mean_absolute_error(y_test, predictions))
print('MSE:', mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(mean_squared_error(y_test, predictions)))
```

```python
sns.distplot(y_test - predictions, bins = 50)
```

```python
#SVR model result : 
# MAE: 0.61067591874093
# MSE: 0.6789591017709762
# RMSE: 0.8239897461564532
```

```python
sns.distplot(y_test - predictions, bins = 50)
```

```python
# SO we know that linear regression is slightly better than SVR, but still not good. Try ANN.
```

```python
# First, we scale our train and test data.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape, X_test.shape
```

```python
# Applying ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam

model = Sequential()

model.add(Dense(6, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))

model.add(Dense(1))
```

```python
# First, we try the rmsprop optimizer, with EarlyStopping but without dropout layers.
model.compile(optimizer = 'rmsprop', loss = 'mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
earlyStop = EarlyStopping(monitor='val_loss', mode = 'min', verbose = 1, patience=20)
model.fit(x = X_train, 
          y = y_train.values, 
          validation_data=(X_test, y_test.values),
          batch_size = 64, 
          epochs=500,
          verbose = 1,
          callbacks = [earlyStop])
```

```python
modelLoss = pd.DataFrame(model.history.history)
modelLoss.plot()
```

```python
# model evaluation:
trainScore = model.evaluate(X_train, y_train, verbose = 0)
testScore = model.evaluate(X_test, y_test, verbose = 0)
trainScore, testScore
```

```python
# the MSEs looking not ideal. Further evaluation:
prediction = model.predict(X_test)
```

```python
predDF = pd.DataFrame(y_test.values, columns=['Y Test'])
prediction = pd.Series(prediction.values.reshape(273,))
predDF = pd.concat([predDF, prediction], axis = 1)
predDF.columns = ['Y Test','Model Predictions']
```

```python
sns.scatterplot(x='Y Test', y = 'Model Predictions', data = predDF)
```

```python
sns.distplot(predDF['Y Test'] - predDF['Model Predictions'], bins = 50)
```

```python
# MAE, MSE, RMSE
print(mean_absolute_error(predDF['Y Test'],predDF['Model Predictions']))
print(mean_squared_error(predDF['Y Test'],predDF['Model Predictions']))
print(np.sqrt(mean_squared_error(predDF['Y Test'],predDF['Model Predictions'])))
```

```python
# Not quite good. Adding dropout layers into ANN.
# Applying ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam

model = Sequential()

model.add(Dense(6, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(6, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(6, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(6, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(1))

model.compile(optimizer = 'rmsprop', loss = 'mse')
```

```python
earlyStop = EarlyStopping(monitor='val_loss', mode = 'min', verbose = 1, patience=20)
model.fit(x = X_train, 
          y = y_train.values, 
          validation_data=(X_test, y_test.values),
          batch_size = 64, 
          epochs=500,
          verbose = 1,
          callbacks = [earlyStop])
```

```python
modelLoss = pd.DataFrame(model.history.history)
modelLoss.plot()
```

```python
# model evaluation:
trainScore = model.evaluate(X_train, y_train, verbose = 0)
testScore = model.evaluate(X_test, y_test, verbose = 0)
trainScore, testScore
```

```python
prediction = model.predict(X_test)
predDF = pd.DataFrame(y_test.values, columns=['Y Test'])
prediction = pd.Series(prediction.reshape(273,))
predDF = pd.concat([predDF, prediction], axis = 1)
predDF.columns = ['Y Test','Model Predictions']
sns.scatterplot(x='Y Test', y = 'Model Predictions', data = predDF)
```

```python
sns.distplot(predDF['Y Test'] - predDF['Model Predictions'], bins = 50)
```

```python
# MAE, MSE, RMSE
print(mean_absolute_error(predDF['Y Test'],predDF['Model Predictions']))
print(mean_squared_error(predDF['Y Test'],predDF['Model Predictions']))
print(np.sqrt(mean_squared_error(predDF['Y Test'],predDF['Model Predictions'])))
```

```python
# So the dropout layers are bad. We now try adam.
model = Sequential()

model.add(Dense(6, activation = 'relu'))

model.add(Dense(6, activation = 'relu'))

model.add(Dense(6, activation = 'relu'))

model.add(Dense(6, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
earlyStop = EarlyStopping(monitor='val_loss', mode = 'min', verbose = 1, patience=20)
model.fit(x = X_train, 
          y = y_train.values, 
          validation_data=(X_test, y_test.values),
          batch_size = 64, 
          epochs=500,
          verbose = 1,
          callbacks = [earlyStop])
```

```python
modelLoss = pd.DataFrame(model.history.history)
modelLoss.plot()
```

```python
# model evaluation:
trainScore = model.evaluate(X_train, y_train, verbose = 0)
testScore = model.evaluate(X_test, y_test, verbose = 0)
trainScore, testScore
```

```python
prediction = model.predict(X_test)
predDF = pd.DataFrame(y_test.values, columns=['Y Test'])
prediction = pd.Series(prediction.reshape(273,))
predDF = pd.concat([predDF, prediction], axis = 1)
predDF.columns = ['Y Test','Model Predictions']
sns.scatterplot(x='Y Test', y = 'Model Predictions', data = predDF)
```

```python
sns.distplot(predDF['Y Test'] - predDF['Model Predictions'], bins = 100)
```

```python
# MAE, MSE, RMSE
print(mean_absolute_error(predDF['Y Test'],predDF['Model Predictions']))
print(mean_squared_error(predDF['Y Test'],predDF['Model Predictions']))
print(np.sqrt(mean_squared_error(predDF['Y Test'],predDF['Model Predictions'])))
```

```python
# Now we perform a PCA analysis comparing to visualize df.
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('LC50 [-LOG(mol/L)]', axis =1))
scaled_data = scaler.transform(df.drop('LC50 [-LOG(mol/L)]', axis =1))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
```

```python
df_comp = pd.DataFrame(pca.components_, columns=df.drop('LC50 [-LOG(mol/L)]', axis =1).columns)
```

```python
plt.figure(figsize = (12, 6))
sns.heatmap(df_comp, cmap='plasma')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],df['LC50 [-LOG(mol/L)]'], cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
corrDF = pd.DataFrame([df['LC50 [-LOG(mol/L)]'], x_pca[:,0]]).transpose()
corrDF.columns = ['LC50 [-LOG(mol/L)]', 'pca indicator']
```

```python
corrDF.corr()
```

```python
corrDF.plot()
```

```python
sns.distplot(corrDF['LC50 [-LOG(mol/L)]'] - corrDF['pca indicator'])
```
